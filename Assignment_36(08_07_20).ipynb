1. From the below inputs and weights calculate the outputs of nodes o1 and o2?

```
i1 = 0.05
i2 = 0.10
w1 = 0.15
w2 = 0.20
w3 = .25
w4 = .30
b1 = .35
# we need to calculate h1,h2
```

```
import numpy as np 
a1 = i1*w1 + i2*w3 + b1
a2 = i1*w2 + i2*w4 + b1
h1 = 1/(1+np.exp(a1))
h2 = 1/(1+np.exp(a2))
w5 = 0.40
w6 = 0.45
w7 = 0.50
w8 = 0.55
b2 = 0.60
o1 = 1/(1+np.exp((h1*w5+h2*w7+b2)))
o2 = 1/(1+np.exp((h1*w6+h2*w8+b2)))
print("o1=",o1)
print("o2=",o2)
```
