5.Apply the KNN Algorithm to the dataset without applying scaling operation and try to compare the results before scaling and after scaling.



```
# Preparing input and output data
X=df.drop(['diagnosis'],axis=1)
y=df.diagnosis
```

```
#splitting the data into testing and training
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

- **Unscaled data**

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
score = []
for k in range(1,21):
  knn_model = KNeighborsClassifier(n_neighbors=k)
  knn_model.fit(X_train,y_train)
  y_pred_test = knn_model.predict(X_test)
  score.append(accuracy_score(y_test,y_pred_test))
  print(accuracy_score(y_test,y_pred_test))
```

```
# visualizing scores to get optimum k value
plt.plot(score,marker="o",linestyle="--",markerfacecolor="red")
plt.grid()
plt.show()
```

  -  Optimum k value is 8

```
final_model = KNeighborsClassifier(n_neighbors=8,metric="euclidean")
final_model.fit(X_train,y_train)
```

```
y_train_pred = final_model.predict(X_train)
y_test_pred=final_model.predict(X_test)
```

  - **Scaled data**

```
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaled_X_train=pd.DataFrame(data=sc.fit_transform(X_train),columns=X_train.columns)
scaled_X_test=pd.DataFrame(data=sc.fit_transform(X_test),columns=X_test.columns)
```

```
#find optimum k value
scores = []
for k in range(1,21):
  knn_model1 = KNeighborsClassifier(n_neighbors=k)
  knn_model1.fit(scaled_X_train,y_train)
  scaled_pred_test = knn_model1.predict(scaled_X_test)
  scores.append(accuracy_score(y_test,scaled_pred_test))
  print(accuracy_score(y_test,scaled_pred_test))
```

```
# visualizing scores to get optimum k value
plt.plot(scores,marker="o",linestyle="--",markerfacecolor="red")
plt.grid()
plt.show()
```

optimum k value is 7

```
scaled_final_model=KNeighborsClassifier(n_neighbors=7,metric='euclidean')
scaled_final_model.fit(scaled_X_train,y_train)
```

```
scaled_y_train_pred=scaled_final_model.predict(scaled_X_train)
scaled_y_test_pred=scaled_final_model.predict(scaled_X_test)
```

```
#confusion matrix before scaling
sns.heatmap(confusion_matrix(y_test,y_test_pred),annot=True,fmt="0.1f")
plt.show()
```

```
print('CLASSIFICATION REPORT before SCALING:\n')
print('\non TESTING DATA:\n')
print(classification_report(y_test,y_test_pred))
```

```
#confusion matrix after scaling
sns.heatmap(confusion_matrix(y_test,scaled_y_test_pred),annot=True,fmt="0.1f")
plt.show()
```

```
print('\nCLASSIFICATION REPORT after SCALING:\n')
print('\non TESTING DATA:\n')
print(classification_report(y_test,scaled_y_test_pred))
```

**OBSERVATION**

- Before scaling accuracy was 94%
- After scaling accuracy was 96%
-  We can say that the model that was trained on scaled values worked much better than the one that was trained on non-uniform values.
- This indicates that, distance based algorithms like KNN produce better outcomes when provide with scalled data, rather than test data
